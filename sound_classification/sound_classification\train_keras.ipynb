```python
import glob
import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

def one_hot_encode(labels):
    n_labels = len(labels)
    n_unique_labels = 10 # len(np.unique(labels))
    one_hot = np.zeros((n_labels,n_unique_labels))
    one_hot[np.arange(n_labels), labels] = 1
    return one_hot
```

```python
fp = open('train_features.csv')
data = [line.strip().split(',') for line in fp]
fp.close()

files = [row[0] for row in data]
labels = [int(row[1]) for row in data]
features = [list(map(float, row[2:])) for row in data]  # X
labels = np.array(labels)
features = np.array(features)

print('Audio Data Loading Done')
print("feature. shape",features.shape)
print("label. shape",labels.shape)
```

```python
labels = one_hot_encode(labels)
for f, l in zip(files, labels):
    print(f, l)
```

```python
# 85% train, 15% validation
selected = np.random.rand(len(features)) < 0.85
train_x = features[selected]
train_y = labels[selected]
valid_x = features[~selected]
valid_y = labels[~selected]

print(selected[:20])
n_dim = features.shape[1]
print('DIM', n_dim)
n_classes = 10

print('학습데이터수', train_x.shape[0])
print('검증데이터수', valid_x.shape[0])
```

```python
from keras.models import Sequential
from keras.layers import Dense
model = Sequential()
model.add(Dense(300, input_dim=n_dim, activation='sigmoid'))
model.add(Dense(200, activation='tanh'))
model.add(Dense(100, activation='sigmoid'))
model.add(Dense(n_classes, activation='softmax'))
model.compile(optimizer='Adam', loss='categorical_crossentropy')
hist = model.fit(train_x, train_y, epochs=400, batch_size=20, 
                 validation_data = (valid_x, valid_y),
                 shuffle=True, verbose=1)
```

```python
model.save('save_keras')
```

```python
# 학습 결과를 하나 하나 확인해 보기 위해
y_pred = model.predict(train_x)
for y_pr, y_real in zip(y_pred, train_y):
    print(' '.join(['%.3f' % y for y in y_pr]), '|', np.argmax(y_pr), np.argmax(y_real))
```

```python
# 검증 결과를 하나 하나 확인해 보기 위해
y_pred = model.predict(valid_x)
for y_pr, y_real in zip(y_pred, valid_y):
    print(' '.join(['%.3f' % y for y in y_pr]), '|', np.argmax(y_pr), np.argmax(y_real))
```
