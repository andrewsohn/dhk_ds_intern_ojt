## 동일 평균값, 분산값을 갖는 4가지 데이터셋
### Anscombe's quartet

```python
import math
import matplotlib.pyplot as plt

def calc_variance_covariance(X, Y):
  assert len(X) == len(Y)
  num = len(X)
  x_mean = sum(X) / num
  y_mean = sum(Y) / num
  x_var = sum([(x - x_mean) * (x - x_mean) for x in X]) / num
  y_var = sum([(y - y_mean) * (y - y_mean) for y in Y]) / num
  cov = sum([(y - y_mean) * (x - x_mean) for x, y in zip(X, Y)]) / num
  return x_var, y_var, cov

  
def load_data(fname):
  data = []
  fp = open(fname)
  for line in fp:
    flds = line.strip().split()
    numvals =  list(map(float, flds))
    data.append( numvals )
  
  return data
```

```python
mydata = load_data('data.csv')
for row in mydata:
    print(row)
```

```python
X1 = [row[0] for row in mydata] # Col 1
Y1 = [row[1] for row in mydata] # Col 2
X2 = [row[2] for row in mydata] # Col 3
Y2 = [row[3] for row in mydata] # Col 4
X3 = [row[4] for row in mydata] # Col 5
Y3 = [row[5] for row in mydata] # Col 6
X4 = [row[6] for row in mydata] # Col 7
Y4 = [row[7] for row in mydata] # Col 8
```

```python
print(X1)
print(Y1)
print(X2)
print(Y2)
print(X3)
print(Y3)
print(X4)
print(Y4)
```

```python
print('X1,Y1', calc_variance_covariance(X1, Y1))
print('X2,Y2', calc_variance_covariance(X2, Y2))
print('X3,Y3', calc_variance_covariance(X3, Y3))
print('X4,Y4', calc_variance_covariance(X4, Y4))
```

```python
# 한 차트에 모두 그려보기
def draw_lines(xypairs, title):
  fig, ax = plt.subplots()  
  for X, Y, label in xypairs:
    ax.scatter(X, Y, label=label)
  ax.set_xlabel('X')
  ax.set_ylabel('Y')
  ax.set_title(title)
  plt.legend()
  plt.show()

##
xypairs = [ (X1, Y1, '1'), 
            (X2, Y2, '2'), 
            (X3, Y3, '3'), 
            (X4, Y4, '4')  ]

draw_lines(xypairs, 'Anscombe')
```

```python
# 각각 그려보기
def draw_x_y(X, Y):
  fig, ax = plt.subplots()
  pair = list(zip(X, Y))
  pair.sort()
  X = [p[0] for p in pair]
  Y = [p[1] for p in pair]
  ax.plot(X, Y)
  ax.set_xlabel('X')
  ax.set_ylabel('Y')
  ax.scatter(X, Y)
  plt.show()

    
    
## 각각 그려보기
draw_x_y(X1, Y1)
draw_x_y(X2, Y2)
draw_x_y(X3, Y3)
draw_x_y(X4, Y4)
```
